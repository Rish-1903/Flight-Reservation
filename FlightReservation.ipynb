```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('flights.csv')
```

```python
df.head()
```

```python
df['Year']=df['Date_of_Journey'].str.split('-').str[0]
df['Month']=df['Date_of_Journey'].str.split('-').str[1]
df['Day']=df['Date_of_Journey'].str.split('-').str[2]
```

```python
df.head()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Year']=df['Year'].astype(int)
df['Month']=df['Month'].astype(int)
df['Day']=df['Day'].astype(int)
```

```python
df.info()
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4})
```

```python
df['Total_Stops'].fillna(1,inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
df['Duration_Hour']=df['Duration'].str.split(' ').str[0]
df['Duration_min']=df['Duration'].str.split(' ').str[1]
```

```python
df['Duration_Hour']=df['Duration_Hour'].str.split('h').str[0]
```

```python
df['Duration_min']=df['Duration_min'].str.split('m').str[0]
```

```python
df.head()
```

```python
df.info()
```

```python
df['Duration_min'].fillna(0,inplace=True)
```

```python
df.info()
```

```python
df[df['Duration_Hour']=='5m']
```

```python
df['Duration_Hour'].replace('5m', 0, inplace=True)
```

```python
df[df['index']==5975]
```

```python
df.loc[df['index'] == 5975, 'Duration_min'] = 5
```

```python
df[df['index']==5975]
```

```python
df.info()
```

```python
df['Duration_Hour']=df['Duration_Hour'].astype(int)
```

```python
df['Duration_min']=df['Duration_min'].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
df.drop('Duration',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Duration']=np.multiply(df['Duration_Hour'],60)+df['Duration_min']
```

```python
df.head()
```

```python
df.drop(['Duration_Hour','Duration_min'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Dep_Hour']=df['Dep_Time'].str.split(':').str[0]
```

```python
df['Dep_min']=df['Dep_Time'].str.split(':').str[1]
```

```python
df.head()
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Dep_Hour']=df['Dep_Hour'].astype(int)
df['Dep_min']=df['Dep_min'].astype(int)
```

```python
df.info()
```

```python
df.drop('index',axis=1,inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder=LabelEncoder()
```

```python
df['Airline']=labelencoder.fit_transform(df['Airline'])
df['Source']=labelencoder.fit_transform(df['Source'])
df['Destination']=labelencoder.fit_transform(df['Destination'])
df['Additional_Info']=labelencoder.fit_transform(df['Additional_Info'])
```

```python
df.head()
```

```python
df.info()
```

```python
X=df.drop('Price',axis=1)
Y=df['Price']
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=42,test_size=0.2)
```

```python
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model=RandomForestRegressor()
```

```python
model.fit(X_train_scaled,Y_train)
```

```python
predictions=model.predict(X_test_scaled)
```

```python
print(predictions)
```

```python
model.score(X_test_scaled,Y_test)
```

```python
predictions1=model1.predict(X_test_scaled)
```
